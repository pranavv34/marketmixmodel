```python
col_list1=data1['Parameters'].tolist()
col_list1
```

```python
data_beta=data[col_list1].copy()
contrib=pd.DataFrame()

for col in col_list1:
    xbeta=data1[data1.Parameters==col]["coef"]
    data_beta[col]=data_beta[col].apply(lambda x: x**xbeta)
    data_beta[col].replace(to_replace = 0, value = 1, inplace=True)
    contrib[col]=(1-(1/data_beta[col]))
    print(col)
```

```python
contrib['dv']=data['dv']
```

```python
contrib2=contrib.copy()
contrib2['row_tot']=contrib.sum(axis = 1)-contrib['dv']
```

```python
contrib2
```

```python
for i in col_list1:
    contrib2[i]=contrib2[i]*contrib2['dv']/contrib2['row_tot']
```
